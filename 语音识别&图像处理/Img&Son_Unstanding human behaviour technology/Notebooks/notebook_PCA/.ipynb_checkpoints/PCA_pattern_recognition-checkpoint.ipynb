To load all the images from a directory, SizeX and SizeY are the image sizes

```python
def PCA_base_creation(Base_dir_name):
    vect_database = []
    for ii in range(500) :
        try:
            File_name = Base_dir_name + "im" + str(ii).zfill(3) + ".bmp"
            #print(File_name)
            img = mpimg.imread(File_name)
            size = SizeX*SizeY
            #print(size)
            vect = img.reshape(size)
            vect_database.append(vect)
        except:
             break;
    return vect_database
```

##  PCA analysis
Perform a PCA on the base of « yes », project one image of « yes » database on the
eigenvectors, retain three projection coefficients, rebuild the image from those
coefficients and analyze the result. Do the same thing retaining all the coefficients.

```python
# -*- coding: utf-8 -*-

from sklearn.decomposition import PCA    # inital reduction
from sklearn.manifold import TSNE                   # final reduction

# img_right_no = mpimg.imread(Base_dir_no + "im000.bmp")
# img_left_no = mpimg.imread(Base_dir_no + "im019.bmp")
# img_up_yes = mpimg.imread(Base_dir_yes + "im017.bmp")
# img_down_yes = mpimg.imread(Base_dir_yes + "im035.bmp")
# img_central_no = mpimg.imread(Base_dir_no + "im010.bmp")
# img_central_yes = mpimg.imread(Base_dir_yes + "im007.bmp")

Base_dir_yes = "./data/yes/"
Base_dir_no = "./data/no/"

BaseTrain_yes = PCA_base_creation(Base_dir_yes)
BaseTrain_no = PCA_base_creation(Base_dir_no)
#BaseTest = PCA_base_creation(Base_dir_no)

def PCA_image(num_dimensions,BaseTrain,NbIm,plot = True):
    # number of Eigen vectors you are going to evaluate
    PCA_img = PCA(n_components=num_dimensions) # 10 dimensions
    PCA_img.fit(BaseTrain)
    #fit()可以说是scikit-learn中通用的方法，每个需要训练的算法都会有fit()方法，它其实就是算法中的“训练”这一步骤。

    # display the mean
    MeanVectYes = PCA_img.mean_
    # display the eigen vectors
    EigenVect = PCA_img.components_

    #Vect2display = EigenVect[0] + MeanVect
    """
    Vect2display = EigenVect[0]
    Eigen_image = Vect2display.reshape(SizeX, SizeY)
    fig = plt.figure() # create a new figure window
    img_grid = fig.add_subplot(1, 1, 1)
    img_grid.imshow(Eigen_image, cmap='gray') 
    plt.show()
    """
    
    ImgToProject = BaseTrain[NbIm].reshape(SizeX, SizeY)
    Vect_projected = PCA_img.transform(BaseTrain)
    #transform: 将数据X转换成降维后的数据
    #Vect_projected = PCA_img.transform(np.transpose(VectToProject))
    reconstructVect = PCA_img.inverse_transform(Vect_projected)
    #inverse_transform: 将降维后的数据转换成原始数据

    Image_reconstr = reconstructVect[NbIm, :].reshape(SizeX, SizeY)
    if plot :
        fig = plt.figure() # create a new figure window
        img_grid = fig.add_subplot(2, 1, 1)
        plt.title("original")
        img_grid.imshow(ImgToProject, cmap='gray')
        img_grid = fig.add_subplot(2, 1, 2)
        plt.title("reconstructed")
        img_grid.imshow(Image_reconstr, cmap='gray')

        plt.show()
    
    return ImgToProject, Image_reconstr

num_dimensions = 10
BaseTrain = BaseTrain_yes
NbIm = 20 #choose an image

PCA_image(num_dimensions,BaseTrain,NbIm)
```

## reconstruction quality measuring
Calculate an **average projection error** (Euclidean distance, max of the absolute value
the error on an image) between the images and the reconstruction of the projected
images in the base of the eigenvectors. 
Use this error to **measure the quality of
reconstruction** when you learn on a basis (yes for example) and you test projection 
reconstruction on a different basis (no). 
Learn on a base and test on the same base.
Display some reconstructed images and compare them to the original images to subjectively measure the reconstruction. Conclude.

```python
from sklearn.metrics.pairwise import euclidean_distances

def av_proj_error(ImgToProject,Image_reconstr):
    dist = euclidean_distances(ImgToProject,Image_reconstr)
    #print(dist.shape,dist)
    error = np.sum(dist)
    #print(error)
    
    return error

def PCA_qual_reconstr(num_dimensions,BaseTrain,NbIm,plot=True):
    ImgToProject, Image_reconstr = PCA_image(num_dimensions,BaseTrain,NbIm,plot)
    qual_reconstr = av_proj_error(ImgToProject,Image_reconstr)
    
    return qual_reconstr
    
num_dimensions = 10
BaseTrain = BaseTrain_yes
NbIm = 20 #choose an image 

for NbIm in range(10):
    PCA_qual_reconstr(num_dimensions,BaseTrain,NbIm,False)
```

## k & quality
Vary the number of eigenvectors($k$) selected and measure the impact on the
reconstruction error and the subjective quality of the reconstruction. Display the
results as a graph.

```python
#num_dimensions = k
BaseTrain = BaseTrain_yes
NbIm = 20 #choose an image 
error_array = []
K = 35
for k in range(K):
    error = PCA_qual_reconstr(k,BaseTrain,NbIm,False)
    error_array.append(error)
#print(error_array)

plt.plot(np.arange(K),error_array)
plt.xlabel('the number of eigenvectors')
plt.ylabel('quality of reconstruction')
plt.show()
```
