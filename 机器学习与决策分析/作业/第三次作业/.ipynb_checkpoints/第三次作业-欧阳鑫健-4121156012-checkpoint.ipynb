## 概率潜在语义分析 PLSA

### 创建PLSA类
[代码参考](https://www.pythonheidong.com/blog/article/715324/28b1b02ccfbd86eb09a1/)

[jieba分词](https://www.oschina.net/p/jieba/?fromerr=LRXZzk9z)

[sklearn vectorizer生成文档的词频向量](https://blog.csdn.net/weixin_43439235/article/details/104962824)

```python
class PLSA:
    def __init__(self,text_list,k,stopwords,stopwords_english):
        self.K = k #k为话题数
        self.text_list = text_list
        self.text_num = len(text_list)
        self.stopwords = stopwords
        self.get_X() 

    def get_X(self):
        self.cuted_text = [jieba.lcut(text) for text in self.text_list] # 四种分词模式：默认是精确模式 
        #self.cuted_text = [jieba.lcut_for_search(text) for text in self.text_list] # 搜索引擎模式
        #ieba.lcut 直接生成的就是一个list
        #jieba.cut生成的是一个生成器，generator，也就是可以通过for循环来取里面的每一个词。
        # 文本信息去处理，去掉停用词和标点符号等
        for i in range(self.text_num):
            for j in range(len(self.cuted_text[i])-1, -1, -1):
                if self.cuted_text[i][j] in stopwords or self.cuted_text[i][j] in stopwords_english:
                    del self.cuted_text[i][j]
        #print(self.cuted_text[0])
        
        self.word_all = [] # 单词集合 W
        for i in self.cuted_text:
            self.word_all.extend(i)
        self.word_set = list(set(self.word_all))
        self.word_num = len(self.word_set)
        self.word_dict = {}
        for index,word in enumerate(self.word_set):
            self.word_dict[word] = index
        self.X = np.zeros((self.word_num, self.text_num))
        for i in range(self.text_num):
            count_ = collections.Counter(self.cuted_text[i])
            for k, v in count_.items():
                self.X[self.word_dict[k],i] = v # 单词k在文本i中出现v次
    
    # 生成模型的对数极大似然   
    def loglikelihood_SC(self):
        llk = 0
        for i in range(self.word_num):
            for j in range(self.text_num):
                llk = llk + self.X[i,j]*np.log(np.sum(self.w_z[i,:]*self.z_d[:,j]))
        return llk

    #共现模型的对数极大似然                  
    def loglikelihood_GX(self):
        llk = 0
        #print(np.shape(self.z), np.shape(self.w_z), np.shape(self.d_z))
        for i in range(self.word_num):
            for j in range(self.text_num):
                llk = llk + self.X[i,j]*np.log(np.sum([self.z[:]*self.w_z[i,:]*self.d_z[j,:]]))
        return llk

    # E-step
    def update_p_z_wd_SC(self):
        self.z_wd  = np.zeros((self.word_num,self.text_num,self.K))
        for i in range(self.word_num):
            for j in range(self.text_num):
                self.z_wd[i,j] = np.array([self.w_z[i]*self.z_d[:,j]]) / np.sum([self.w_z[i]*self.z_d[:,j]])

    def update_p_z_wd_GX(self):
        self.z_wd  = np.zeros((self.word_num,self.text_num,self.K))
        for i in range(self.word_num):
            for j in range(self.text_num):
                self.z_wd[i,j] = np.array([self.z[:]*self.d_z[j,:]*self.w_z[i,:]]) / np.sum([self.z[:]*self.d_z[j,:]*self.w_z[i,:]])

    # M-step
    def fit(self,max_iter=20, shengcheng = 1, gongxian = 0):
        if shengcheng:
            self.w_z  = np.random.random((self.word_num, self.K)) # Return random floats in the half-open interval [0.0, 1.0).
            self.z_d = np.random.random((self.K, self.text_num))
            llk_temp = float('-inf')
            for iter in range(max_iter):
                d = np.abs(llk_temp - self.loglikelihood_SC())
                print('生成模型， step:',iter, '\t', d)
                if d > 1e-2:
                    llk_temp = self.loglikelihood_SC()
                    self.update_p_z_wd_SC()
                    for k in range(self.K):
                        for i in range(self.word_num):
                            self.w_z[i,k] = np.sum(self.X[i]*self.z_wd[i,:,k])/\
                            np.sum(self.X*self.z_wd[:,:,k])
                        for j in range(self.text_num):
                            self.z_d[k,j] = np.sum(self.X[:,j]*self.z_wd[:,j,k])/np.sum(self.X[:,j])

        if gongxian:
            self.w_z = np.random.random((self.word_num, self.K))
            self.d_z = np.random.random((self.text_num, self.K))
            self.z = np.random.random(self.K)
            llk_temp = float('-inf')
            for iter in range(max_iter):
                d = np.abs(llk_temp - self.loglikelihood_GX())
                print('共现模型， step: ',iter,'\t', d)
                if  d > 1e-2:
                    llk_temp = self.loglikelihood_GX()
                    self.update_p_z_wd_GX()
                    for k in range(self.K):
                        for i in range(self.word_num):
                            self.w_z[i,k] = np.sum(self.X[i]*self.z_wd[i,:,k])/\
                            np.sum(self.X*self.z_wd[:,:,k])

                        for j in range(self.text_num):
                            self.d_z[j,k] = np.sum(self.X[:,j]*self.z_wd[:,j,k])/\
                            np.sum(self.X*self.z_wd[:,:,k])

                        self.z[k] = np.sum(self.X*self.z_wd[:,:,k])/np.sum(self.X)
```

### PLSA + Einstein求和

参考：[如何理解和使用NumPy.einsum？](https://zhuanlan.zhihu.com/p/27739282)

```python
class PLSA_einsum:
    def __init__(self,text_list,k,stopwords,stopwords_english):
        self.K = k #k为话题数
        self.text_list = text_list
        self.text_num = len(text_list)
        self.stopwords = stopwords
        self.get_X() 

    # 获得文本-单词矩阵
    def get_X(self): 
        #self.cuted_text = [jieba.lcut(text) for text in self.text_list] # 四种分词模式：默认是精确模式 
        self.cuted_text = [jieba.lcut_for_search(text) for text in self.text_list] # 搜索引擎模式
        #ieba.lcut 直接生成的就是一个list
        #jieba.cut生成的是一个生成器，generator，也就是可以通过for循环来取里面的每一个词。
        # 文本信息去处理，去掉停用词和标点符号等
        for i in range(self.text_num):
            for j in range(len(self.cuted_text[i])-1, -1, -1):
                if self.cuted_text[i][j] in stopwords or self.cuted_text[i][j] in stopwords_english:
                    del self.cuted_text[i][j]
        #print(self.cuted_text[0])
        self.word_all = [] # 单词集合 W
        for i in self.cuted_text:
            self.word_all.extend(i)
        self.word_set = list(set(self.word_all))
        self.word_num = len(self.word_set)
        self.word_dict = {}
        for index,word in enumerate(self.word_set):
            self.word_dict[word] = index
        self.X = np.zeros((self.word_num, self.text_num))
        for i in range(self.text_num):
            count_ = collections.Counter(self.cuted_text[i])
            for k, v in count_.items():
                self.X[self.word_dict[k],i] = v # 单词k在文本i中出现v次

    # 生成模型的对数极大似然   
    def loglikelihood_SC(self):
        c1 = np.log(np.einsum('ik, kj->ij', self.w_z, self.z_d))
        llk = np.einsum('ij, ij ->', self.X, c1)
        return llk

    #共现模型的对数极大似然                  
    def loglikelihood_GX(self):
        c1 = np.log(np.einsum('k, ik, jk->ij', self.z, self.w_z, self.d_z))
        llk = np.einsum('ij, ij ->', self.X, c1)
        return llk

    # E-step
    def update_p_z_wd_SC(self):
        self.z_wd  = np.zeros((self.K, self.word_num,self.text_num))
        numerator = np.einsum('ik, kj->kij', self.w_z, self.z_d)
        denominator = np.einsum('kij->ij', numerator)
        #print(np.shape(numerator), np.shape(denominator)) 
        denominator = np.expand_dims(denominator, 0).repeat(self.K, axis=0)
        denominator[denominator == 0] = 1
        #print(np.shape(denominator)) 
        self.z_wd = numerator / denominator

    def update_p_z_wd_GX(self):
        self.z_wd  = np.zeros((self.K, self.word_num,self.text_num))
        numerator = np.einsum('k, ik, jk ->kij', self.z, self.w_z, self.d_z)
        denominator = np.einsum('kij->ij', numerator)
        #print(np.shape(numerator), np.shape(denominator)) 
        denominator = np.expand_dims(denominator, 0).repeat(self.K, axis=0)
        denominator[denominator == 0] = 1
        #print(np.shape(denominator)) 
        self.z_wd = numerator / denominator

    # EM algo and iterations
    def fit(self,max_iter=20, shengcheng = 1, gongxian = 0):
        if shengcheng:
            self.w_z  = np.random.random((self.word_num, self.K)) # Return random floats in the half-open interval [0.0, 1.0).
            self.z_d = np.random.random((self.K, self.text_num))
            llk_temp = float('-inf')
            for iter in range(max_iter):
                d = np.abs(llk_temp - self.loglikelihood_SC())
                if d > 1e-2:
                    print('生成模型， step:',iter, '\t', d)
                    llk_temp = self.loglikelihood_SC()
                    # E-step
                    self.update_p_z_wd_SC() 
                    # M-step
                    # update p_w_z
                    numerator = np.einsum('ij, kij->ik', self.X, self.z_wd)
                    denominator = np.einsum('mj, kmj->k', self.X, self.z_wd)
                    denominator = np.expand_dims(denominator, 0).repeat(self.word_num, axis = 0)
                    denominator[denominator == 0] = 1
                    self.w_z = numerator / denominator
                    # update p_z_d
                    numerator = np.einsum('ij, kij->kj', self.X, self.z_wd)
                    denominator = np.einsum('ij->j',self.X)
                    denominator = np.expand_dims(denominator, 0).repeat(self.K, axis = 0)
                    denominator[denominator == 0] = 1
                    self.z_d = numerator / denominator
                else:
                    print('<1e-2, iteration ends')
                    break

        if gongxian:
            self.w_z = np.random.random((self.word_num, self.K))
            self.d_z = np.random.random((self.text_num, self.K))
            self.z = np.random.random(self.K)
            llk_temp = float('-inf')
            for iter in range(max_iter):
                d = np.abs(llk_temp - self.loglikelihood_GX())
                if  d > 1e-2:
                    print('共现模型， step: ',iter,'\t', d)
                    llk_temp = self.loglikelihood_GX()
                    # E-step
                    self.update_p_z_wd_GX()
                    # M-step
                    # update p_z
                    numerator = np.einsum('ij, kij->k', self.X, self.z_wd)
                    denominator = np.einsum('ij->', self.X)
                    #print(np.shape(numerator), np.shape(denominator)) 
                    denominator = np.expand_dims(denominator, 0).repeat(self.K, axis = 0)
                    denominator[denominator == 0] = 1
                    #print(np.shape(denominator)) 
                    self.z = numerator / denominator
                    # update p_w_z
                    numerator = np.einsum('ij, kij->ik', self.X, self.z_wd)
                    denominator = np.einsum('mj, kmj->k', self.X, self.z_wd)
                    denominator = np.expand_dims(denominator, 0).repeat(self.word_num, axis = 0)
                    denominator[denominator == 0] = 1
                    self.w_z = numerator / denominator
                    # update p_d_z
                    numerator = np.einsum('ij, kij->jk', self.X, self.z_wd)
                    denominator = np.einsum('in, kin->k', self.X, self.z_wd)
                    denominator = np.expand_dims(denominator, 0).repeat(self.text_num, axis = 0)
                    denominator[denominator == 0] = 1
                    self.d_z = numerator / denominator
                else:
                    print('<1e-2, iteration ends')
                    break
```

#### 生成模型

```python
from string import punctuation

#文本信息预处理-stopwords
stopwords = [i.strip() for i in open('cn_stopwords.txt').readlines()] 
stopwords_english = ['of','edu','web','org','doi','2021','2020','qq','weixin','for',\
                     'www','and','content','cn','html','com','https','http','gov',\
                    'xjtu', 'Three', 'chromosome', '10','IPv6',',','.','Science',\
                    '2018', '（', '）', 'b', '=']  

text_list = np.load("Content.npy")
K = 5 # 话题数
lsa1 = PLSA_einsum(text_list, k = K, stopwords = stopwords, stopwords_english = stopwords_english)
#lsa1 = PLSA(text_list, k = K, stopwords = stopwords, stopwords_english = stopwords_english)

print(lsa1.cuted_text[19])
```

**输出不同的话题数下各个话题的高频词**

```python
# print('total number of words: ',np.shape(lsa.word_set))
# print('w_z:\n',np.shape(lsa.w_z))#, lsa.w_z)
# print('z_d:\n',np.shape(lsa.z_d))#,lsa.z_d)

p_w_z = lsa1.w_z
word_dict = lsa1.word_dict

T = 10 #前5个高频词
def get_key(dict, value):
    return [k for k, v in dict.items() if v == value]
    
topic_list = [ ]

for k in range(K):
    top_index = p_w_z[:,k].argsort()[-T:] 
    #print(top_index)
    topics = list(get_key(word_dict,i) for i in top_index)
    print(topics,'\n')
    topic_list.append(topics)

#print(topic_list)
```

**Note**

对100条新闻进行概率潜在语义分析，分别建立了它们的生成模型和共现模型。话题数为**5**，每个话题对应的高频词取前**10**个， 不同的话题数下各个话题的高频词如上所示。

当新闻文本D数据越大时，矩阵维度增高，使用循环计算效率低，耗时长。故采用**Einstein求和**方式调用**np.einsum()**函数，从而显著提高运算效率，
