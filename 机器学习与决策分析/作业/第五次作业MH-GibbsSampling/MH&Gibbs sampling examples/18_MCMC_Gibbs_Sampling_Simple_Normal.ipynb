# MCMC Estimation for a Simple Normal Model
This notebook shows how to perform a simple MCMC simulation for a regression model with intercept only. We will use semi-conjugate priors and illustrate how Gibbs sampling can be used to perform simulation based inference for this model .

## Model
$$y_i = \mu + e_i;e_i \sim N(0,\sigma^2)$$
The likelihood for the sample can be written as
$$p\left(y | \mu, \sigma^{2}\right)=\prod_{i=1}^{n} \frac{1}{\sqrt{2 \pi \sigma^{2}}} \exp \left(-\frac{1}{2} \frac{\left(y_{i}-\mu\right)^{2}}{\sigma^{2}}\right)$$
## Priors
We assume independent priors on the two unknowns. Assume that the prior for $\mu$ is a univariate normal given by $p(\mu)=N(\eta,\tau^2)$ and the prior for $\sigma^2$ is inverse gamma given by $p(\sigma^{2})=\mathrm{IG}(a, b)$. The joint prior is then given by $p(\mu,\sigma^2)=p(\mu)p(\sigma^2)$. Notice that the normal prior for $\mu$ is conjugate to the normal likelihood when $\sigma^2$ is known, and similarly, the inverse gamma prior for $\sigma^2$ is conjugate when $\mu$ is known. However, the joint prior is not conjugate.
## Posterior
As the prior $p(\mu, \sigma^2)$ is not conjugate to the likelihood, the posterior cannot be written in closed form. The posterior is given by 

$$
p(\mu, \sigma^2 | y) = \frac{p(y | \mu, \sigma^2)p(\mu) p(\sigma^2)}{\int\int p(y | \mu, \sigma^2) p(\mu) p(\sigma^2) d \mu d \sigma^2}
$$

## Full Conditionals
MCMC methods involve simulating from the full conditionals of the unknowns. We now focus on the full conditionals. 

### Full Conditional for the Normal Mean $\mu$

Suppose $\bar{y}$ is the mean of the dataset. The full conditional for $\mu$ is given by 

$$
p\left(\mu |\alpha ,\left\{y_i\right\}\right)=N\left(\hat{\mu },v_{\mu }\right)
$$

where, $v_{\mu }^{-1}=\tau^{-2}+n\sigma^{-2}$ and $\hat{\mu}=v_{\mu}\left[\tau^{-2}\eta+n\sigma^{-2}\bar{y}\right]$. Simulating $\mu$ involves simulating from the above univariate normal distribution. 

### Full Conditional for the Normal Variance $\sigma^{2}$

This full conditional for $\sigma^2$ is inverse gamma and is given by 

$$
p\left(\sigma^2|\mu ,\left\{y_i\right\}\right)=\mathrm{IG}\left(a_1,b_1\right)
$$

where $a_{1}=a+n/2$ and $b_{1}=\big(\frac{1}{2}\sum_{i=1}^{n}(y_{i}-\mu)^{2}+b^{-1}\big)^{-1}$

```python
import numpy as np
#Functions for generating draws

#The following function generates a random draw from the full conditional 
#distribution of mu. The function argument n, refers to the sample size and ybar refers to the sample mean.
def pmu(eta, tauSq, sigmaSq, ybar, n):
    precision = 1.0 / tauSq + n / sigmaSq
    vmu = 1.0 / precision
    mumean = vmu * ((1.0 / tauSq) * eta + n / sigmaSq * ybar)
    return np.random.normal(mumean, np.sqrt(vmu))
    

#The following function generates a random draw from the full conditional distribution of sigma^2. 
def pSigmaSq(mu, n, yvec, a, b):
    aprime = a+n/2
    bprime = (np.sum((yvec - mu)**2) / 2.0 + 1 / b)
    bprime = 1 / bprime
    return 1.0/np.random.gamma(aprime, bprime)
```

```python
#Generating Data
n = 100
yvec = np.random.normal(5,1,n)
ybar = np.mean(yvec)
std = np.std(yvec)
print(ybar)
print(std)
```

```python
#MCMC Draws
#Starting values
sigmaSq = 1.0; mu = 0.1;niter = 2000
#Priors
eta =0;tauSq=100;a=3;b=1;
#Draws
muDraws = []
sigmaSqDraws = []
for iter in range(niter):
    mu = pmu(eta, tauSq, sigmaSq, ybar, n)
    sigmaSq = pSigmaSq(mu, n, yvec, a, b)
    muDraws.append(mu)
    sigmaSqDraws.append(sigmaSq)
```

```python
import matplotlib.pyplot as plt
plt.subplots(figsize=(20, 10)) 
plt.subplot(1,2,1)
plt.plot(range(niter),np.array(muDraws))
plt.xlabel('mu')
plt.subplot(1,2,2)
plt.plot(range(niter),np.array(sigmaSqDraws))
plt.xlabel('sigma')
plt.show()
```
