### 1.2 数据生成 
假设硬币 A、B、C 正面向上的概率分别是 0.7、0.3、0.6，按照三硬币模型的数据生成过程独立地重复 100 次试验并记录观测结果的序列；


```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
pi = 0.7
p = 0.3
q = 0.6

# p(y|theta)
py0 = pi*(1-p) + (1-pi)*(1-q) # p(y=0)
py1 = pi*p + (1-pi) * q # p(y=1)
print(py0,py1)

#data generating given probabilities
N = 100
np.random.seed(0)
P =[py0, py1]
Y = np.random.choice([0, 1],size = N, p = P)
print('Observations:\n ',Y)
```

### 1.3 利用 EM 算法根据上述观测序列估计各硬币正面向上的概率

```python
def loglikelihood(pi,p,q,Y):
    llh = 0
    for yj in Y:
        llh = llh + np.log(float((pi * math.pow(p, yj) * math.pow(1 - p, 1 - yj) + \
            (1 - pi) * math.pow(q, yj) * math.pow(1 - q, 1 - yj))))
    return llh

def cal_u(pi,p,q,yj):
    # calculate u_j(i+1) for each yj
    return pi * math.pow(p, yj) * math.pow(1 - p, 1 - yj) / \
           float(pi * math.pow(p, yj) * math.pow(1 - p, 1 - yj) +
                 (1 - pi) * math.pow(q, yj) * math.pow(1 - q, 1 - yj))

def E_step(pi,p,q,Y):
    # calculate U = [u_j(i+1)]   
    return [cal_u(pi,p,q,yj) for yj in Y]

def M_step(U,Y):
    N = len(U)
    pi_temp = 1/N * np.sum(U)
    p_temp = U @ Y.T / np.sum(U)
    q_temp = (np.ones(N)-U) @ Y.T / np.sum(np.ones(N)-U)
    return [pi_temp,p_temp,q_temp]
    
def run(Y,pi0,p0,q0,ite = 200, ifprint = False):
    [pi_temp,p_temp,q_temp] = [pi0,p0,q0]
    llh_temp = float('-inf')
    LLH = np.zeros(ite)
    for i in range(ite):
        # 以最大似然函数估计作为判别准则
        if  loglikelihood(pi_temp,p_temp,q_temp,Y) > llh_temp:
            llh_temp = loglikelihood(pi_temp,p_temp,q_temp,Y)
            U = E_step(pi_temp,p_temp,q_temp,Y)
            [pi_temp,p_temp,q_temp] = M_step(U,Y)
            if ifprint:
                print('step',i+1, ':',pi_temp,p_temp,q_temp,'\nlogLikelihood: ', llh_temp)
        else:
            break
        
    return pi_temp,p_temp,q_temp
```

```python
# initiatial values
pi0, p0, q0 = 0.8, 0.25, 0.5
print('initial vaules',pi0, p0, q0)
# hiden parameters estimation
pi_est, p_est, q_est = run(Y,pi0,p0,q0,ifprint=True)
llh = loglikelihood(pi_est,p_est,q_est,Y)
print('true values: ', pi, p, q)
```

**试验不同的初始值组合$(\pi0,p0,q0)$**

```python
Pr = np.arange(0.01, 0.99, 0.05, float) 
Results = np.zeros((len(Pr)**3,4))
i = 0
for pi0 in Pr:
    for p0 in Pr:
        for q0 in Pr:
            pi_est, p_est, q_est = run(Y,pi0,p0,q0)
            llh = loglikelihood(pi_est,p_est,q_est,Y)
            Results[i,:] = [pi_est, p_est, q_est, llh]
            i = i+1
```

```python
print('Available values for pi0, p0 and q0',Pr) 
print('\n\t \t ******** Estimations ********')
print('pi_est, \t p_est, \t q_est, \t loglikelihood')
print(Results)
maxLikelihood = np.argmax(Results[:,3])
pos = np.argmax(Results[:,3], axis = 0)
print('parameters with maximum loglikelihood:\n',Results[pos])
```

对模型参数的初值多次取值，进行试验，**可得EM算法的结果对初值非常敏感**

```python
# means and covars
mu1 = [3, 1]
Sigma1 = [[1, -0.5],[-0.5, 1]]
mu2 = [8, 10]
Sigma2 = [[2, 0.8],[0.8, 2]]
mu3 = [12, 2]
Sigma3 = [[1, 0],[0, 1]]
print(mu1,Sigma1,'\n',mu2,Sigma2,'\n',mu3,Sigma3)
```

### 3.1 Data generating

```python
N= 300
D= 2
x1 = np.zeros((N, D))
x2 = np.zeros((N, D))
x3 = np.zeros((N, D))
for n in np.arange(N):
    x1[n,:] = np.random.multivariate_normal(mu1, Sigma1)
    x2[n,:] = np.random.multivariate_normal(mu2, Sigma2)
    x3[n,:] = np.random.multivariate_normal(mu3, Sigma3)

X = np.zeros((3*N,D))
X = np.vstack((x1,x2,x3))

print(X.shape, x1.shape,x2.shape,x3.shape)
```

#### Ground truth data plotting

```python
fig1 = plt.figure(figsize=(10,4))
ax1 = fig1.add_subplot(111)
ax1.scatter(X[:, 0], X[:, 1], s=10)
ax1.set_title('Ground truth(X dataset)')

fig2 = plt.figure(figsize=(10,4))
ax2 = fig2.add_subplot(111)
ax2.scatter(x1[:, 0], x1[:, 1], s=10, color = 'b')
ax2.scatter(x2[:, 0], x2[:, 1], s=10, color = 'g')
ax2.scatter(x3[:, 0], x3[:, 1], s=10, color = 'r')
ax2.set_title('Ground truth(x1,x2,x3 dataset)')
plt.show()
```

### 3.2 GMM construction

**n_components = 3**

```python
from sklearn.mixture import GaussianMixture

gm = GaussianMixture(n_components=3, covariance_type = 'full')
gm.fit(X)

aic = gm.aic(X)
bic = gm.bic(X)
print(aic,bic)
gm.get_params()
```

#### 学习得到的模型与真实模型的对比

```python
labels = gm.predict(X)

fig3 = plt.figure(figsize=(10,4))
ax3 = fig3.add_subplot(111)
ax3.scatter(X[:, 0], X[:, 1], c=labels, s=10, cmap='viridis')
ax3.set_title('Estimation of GMM(X dataset)')


fig2 = plt.figure(figsize=(10,4))
ax2 = fig2.add_subplot(111)
ax2.scatter(x1[:, 0], x1[:, 1], s=10, color = 'b')
ax2.scatter(x2[:, 0], x2[:, 1], s=10, color = 'g')
ax2.scatter(x3[:, 0], x3[:, 1], s=10, color = 'r')
ax2.set_title('Ground truth(x1,x2,x3 dataset)')
plt.show()
```

### 3.3 AIC/BIC for different **n_compontents**

```python
N_comp = 20 # n_compontents in gmm

AIC = np.zeros((N_comp))
BIC = np.zeros((N_comp))

for i in range(N_comp):
    gm = GaussianMixture(n_components = i+1, covariance_type = 'full')
    gm.fit(X)
    aic = gm.aic(X)
    bic = gm.bic(X)
    AIC[i], BIC[i] = aic, bic
    #print(i, aic, bic)
print('AIC: ',AIC,'\n','BIC: ',BIC)
```

```python
fig4 = plt.figure(figsize=(6,4))

plt4 = fig4.add_subplot(111)
plt4.plot(range(N_comp)+np.ones(N_comp), AIC, label='AIC')
plt4.plot(range(N_comp)+np.ones(N_comp), BIC, label='BIC')
plt4.legend()
plt4.set_title('AIC/BIC - N_comp')
plt.show()
```

**The best n_compontents parameter in GMM is 3** where AIC and BIC are the lowest.
